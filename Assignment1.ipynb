```python
emp_name=input("enter the employee name:")
salary= float(input("enter the salary:"))
exp = int(input("enter the no of years of experience:"))

def add_bonus():
    
    if exp>5 and salary<50000:
        print("add 10% bonus to the salary.")
        return salary + (salary*10/100)
    elif exp>=3 and exp<=5:
        print("add 5% bonus to the salary.")
        return salary + (salary*5/100)
    elif exp<3:
        print("No bonus.")
print("final salary:",add_bonus())
```
