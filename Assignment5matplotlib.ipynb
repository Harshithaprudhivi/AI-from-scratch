```python
import matplotlib.pyplot as plt
```

```python
years=list(range(2008,2026))
runs = [165, 246, 307, 557, 364, 634, 359, 505, 973, 308, 530, 464, 466, 341, 639, 722, 741, 620]
averages = [15.0, 22.3, 27.9, 46.4, 28.0, 45.3, 27.6, 45.0, 81.1, 22.4, 48.3, 33.1, 41.1, 28.4, 53.2, 55.4, 61.7, 48.9]
```

```python
class ViratIPL:
    def __init__(self,years,runs,averages):
        self.years = years
        self.runs = runs
        self.averages = averages
    def plot_performance(self):
        fig,axes=plt.subplots(2,1,figsize=(10,8))
        axes[0].bar(self.years,self.runs)
        axes[0].set_title("virat IPL runs over 2008-2025")
        axes[0].set_xlabel("year")
        axes[0].set_ylabel("runs")

        axes[1].plot(self.years,self.averages,marker='o')
        axes[1].set_title("virat IPL averages over 2008-2025")
        axes[1].set_xlabel("Year")
        axes[1].set_ylabel("Batting Average")
```

```python
        plt.tight_layout()
        plt.show()
```

```python

vk = ViratIPL(years, runs, averages)
vk.plot_performance()
```
