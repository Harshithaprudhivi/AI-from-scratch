```python
#plot a scatterplot to analyze the relationship between unitprice and customer rating based on gender
```

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv("sales_data.csv")
```

```python
df.head()
```

```python
plt.figure()
```

```python
sns.scatterplot(data=df,x="Unit_Price",y="Customer_Rating",hue="Gender")
plt.title("Unit Price vs Customer Rating by Gender")
plt.show()
```

```python
#plot distribution of sales amount
```

```python
plt.figure()
sns.histplot(df["Sales_Amount"],kde=True)
plt.title("Distribution of sales amount")
plt.xlabel("salesamount")
plt.ylabel("Frequency")
plt.show()
```

```python
#show no of purchases made in each region
```

```python
plt.figure()
sns.countplot(data=df,x="Region")
plt.title("Number of Purchases by region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.show()
```

```python
#plot distribution of sales amount for each product
```

```python
plt.figure()
sns.boxplot(data=df,x="Product",y="Sales_Amount")
plt.title("Sales amount ditribution by product")
plt.show()
```
